# Data Visualisation - Line and Scatter Plots

## Install matplotlib
- pip3 install matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(10)
print(x)
```

```python
plt.plot(x)
```

```python
x=x**2
plt.plot(x)
```

```python
plt.plot(x)
plt.show()
```

```python
x=np.arange(10)
x2=x**2
x3=x**3
plt.plot(x)
plt.plot(x2)
plt.plot(x3)
plt.show()
```

```python
plt.plot(x)
plt.show()
plt.plot(x2)
plt.show()
plt.plot(x3)
plt.show()
```

```python
plt.scatter(x,x*2,marker=5,color='orange',label='Stock Price')
#plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Market Price Prediction")
plt.show()
```

```python
plt.scatter(x,x*2,marker='X',color='orange',label='Stock Price')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Market Price Prediction")
plt.show()
```

```python
plt.scatter(x,x*2,marker="$ff$",color='orange',label='Stock Price')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Market Price Prediction")
plt.show()
```

```python
# marker reference : https://matplotlib.org/api/markers_api.html
```

```python
plt.scatter(x,x*2,marker='o',color='orange',label='Stock Price')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Market Price Prediction")
plt.show()
```

```python
plt.plot(x,x*2,marker='3',color='orange',label='Stock Price')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Market Price Prediction")
plt.show()
```

```python
plt.scatter(x,x*2,marker='4',color='orange',label='Stock Price')
plt.legend()
plt.xlabel("Days")
plt.ylabel("Price")
plt.title("Stock Market Price Prediction")
plt.show()
```
