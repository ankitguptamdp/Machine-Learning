```python
im=cv2.imread('../Pictures/elephant.jpg') # Reads as BGR format
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)# Conversion from BGR to RGB format
print(im.shape) # pixels length breadth and depth
original_shape=im.shape
plt.imshow(im) # Displays as RGB format
plt.show()
```

```python
# Flatten Each Channel Of The Image (Red ,Green, Blue)
all_pixels=im.reshape((330*500,3))
print(all_pixels.shape)
```

```python
# Flatten Each Channel Of The Image
all_pixels=im.reshape((-1,3))
print(all_pixels.shape)
# -1 will automatically determine the shape
```

```python
# Flatten Each Channel Of The Image
all_pixels=im.reshape((-1,3)) # 3 is necessary other wise shape will be (495000,) if after , nothing is mentioned
print(all_pixels.shape)
# -1 will automatically determine the shape
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors=4
km=KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)
```

# 02. K-Means - Extracting Color Swatches

```python
centers=km.cluster_centers_
print(centers)
# Three values of each cluster center represents Red , Green and Blue Values
```

```python
import numpy as np
```

```python
# We need to convert it to an interger
centers=np.array(centers,dtype='uint8')
# unsigned integers of 8 bits value from 0 to 255
# 255 = 0b11111111
```

```python
print(centers)
# 1 0 1 is close to black
```

### Plot what all colors are these?

```python
plt.figure?
```

```python
i=1
plt.figure(0,figsize=(4,2))
colors=[]
for each_col in centers:
    plt.subplot(1,4,i)
    i+=1
    colors.append(each_col)
    
    # Color Swatches (Samples)
    a=np.zeros((100,100,3),dtype='uint8')
    a[:,:,:]=each_col
    plt.imshow(a)
plt.show()
```

# 03. K-Means - Image Segmentation

### Segmenting Our Original Image

```python
new_img=np.zeros((330*500,3),dtype='uint8')
print(new_img.shape)
```

```python
print(colors)
```

```python
labels=km.labels_
print(labels)
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix]=colors[labels[ix]]
    
plt.imshow(im)
plt.show()
new_img=new_img.reshape((original_shape))
plt.imshow(new_img)
plt.show()
```
